```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("../../data/kgh_2005_beta.csv")
print(data.info())
```

```python
data['Data'] = pd.to_datetime(data['Data'])
data.set_index('Data', inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
X = data[['Otwarcie', 'Najwyzszy', 'Najnizszy', 'Zamkniecie', 'Wolumen']]
y = data['Beta']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
test_predictions = model.predict(X_test)
mse = np.mean((test_predictions - y_test)**2)
r_squared = model.score(X_test, y_test)
```

```python
print("Mean Squared Error:", mse)
print("R-squared:", r_squared)
```

```python
import matplotlib.pyplot as plt

# Przewidywane bety dla danych testowych
predicted_values = pd.Series(test_predictions, index=X_test.index)

# Rzeczywiste bety
actual_values = y_test

# Tworzymy wykres
plt.figure(figsize=(12, 6))
plt.plot(predicted_values.index, predicted_values, label='Predicted Beta', color='blue')
plt.plot(actual_values.index, actual_values, label='Actual Beta', color='red')
plt.xlabel('Date')
plt.ylabel('Beta')
plt.title('Predicted vs Actual Beta')
plt.legend()
plt.show()
```
